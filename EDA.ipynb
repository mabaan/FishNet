##  Exploratory Data Analysis

Let's explore the PhiUSIIL Phishing URL Dataset to understand the characteristics of phishing vs legitimate URLs.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

### 1. Class Distribution

```python
# Analyze class distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
label_counts = df['label'].value_counts()
axes[0].bar(['Phishing (0)', 'Legitimate (1)'], label_counts.values, color=['#e74c3c', '#2ecc71'])
axes[0].set_ylabel('Count')
axes[0].set_title('URL Class Distribution')
for i, v in enumerate(label_counts.values):
    axes[0].text(i, v + 500, str(v), ha='center', fontweight='bold')

# Pie chart
colors = ['#e74c3c', '#2ecc71']
axes[1].pie(label_counts.values, labels=['Phishing', 'Legitimate'], autopct='%1.1f%%', 
            colors=colors, startangle=90)
axes[1].set_title('URL Class Proportion')

plt.tight_layout()
plt.show()

print(f"Total URLs: {len(df)}")
print(f"Phishing URLs: {label_counts[0]} ({label_counts[0]/len(df)*100:.1f}%)")
print(f"Legitimate URLs: {label_counts[1]} ({label_counts[1]/len(df)*100:.1f}%)")
print(f"Balance Ratio: {label_counts[1]/label_counts[0]:.2f}:1")
```

### 2. URL Length Analysis

```python
# Analyze URL length differences
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Box plot
axes[0].boxplot([df[df['label']==0]['URLLength'], df[df['label']==1]['URLLength']], 
                labels=['Phishing', 'Legitimate'])
axes[0].set_ylabel('URL Length')
axes[0].set_title('URL Length Distribution by Class')
axes[0].grid(True, alpha=0.3)

# Histogram
axes[1].hist(df[df['label']==0]['URLLength'], bins=50, alpha=0.6, label='Phishing', color='#e74c3c')
axes[1].hist(df[df['label']==1]['URLLength'], bins=50, alpha=0.6, label='Legitimate', color='#2ecc71')
axes[1].set_xlabel('URL Length')
axes[1].set_ylabel('Frequency')
axes[1].set_title('URL Length Distribution')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nURL Length Statistics:")
print(df.groupby('label')['URLLength'].describe())
```

### 3. Domain Length Analysis

```python
# Domain length comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].boxplot([df[df['label']==0]['DomainLength'], df[df['label']==1]['DomainLength']], 
                labels=['Phishing', 'Legitimate'])
axes[0].set_ylabel('Domain Length')
axes[0].set_title('Domain Length Distribution by Class')
axes[0].grid(True, alpha=0.3)

axes[1].hist(df[df['label']==0]['DomainLength'], bins=30, alpha=0.6, label='Phishing', color='#e74c3c')
axes[1].hist(df[df['label']==1]['DomainLength'], bins=30, alpha=0.6, label='Legitimate', color='#2ecc71')
axes[1].set_xlabel('Domain Length')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Domain Length Distribution')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nDomain Length Statistics:")
print(df.groupby('label')['DomainLength'].describe())
```

### 4. Security Features Analysis (HTTPS, TLD)

```python
# HTTPS usage analysis
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# HTTPS distribution by class
https_data = df.groupby(['label', 'IsHTTPS']).size().unstack(fill_value=0)
https_data.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#2ecc71'])
axes[0].set_xticklabels(['Phishing', 'Legitimate'], rotation=0)
axes[0].set_ylabel('Count')
axes[0].set_title('HTTPS Usage by URL Class')
axes[0].legend(['No HTTPS', 'HTTPS'], title='Protocol')
axes[0].grid(True, alpha=0.3)

# TLD analysis - top 10
tld_counts = df.groupby(['label', 'TLD']).size().unstack(fill_value=0)
top_tlds = tld_counts.sum(axis=1).nlargest(10).index
tld_top = tld_counts.loc[top_tlds]
tld_top.T.plot(kind='bar', ax=axes[1], color=['#e74c3c', '#2ecc71'])
axes[1].set_ylabel('Count')
axes[1].set_title('Top 10 TLDs by Class')
axes[1].set_xlabel('TLD')
axes[1].legend(['Phishing', 'Legitimate'])
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nHTTPS Usage Statistics:")
print(df.groupby('label')['IsHTTPS'].value_counts(normalize=True).unstack(fill_value=0))
```

### 5. Special Characters Analysis

```python
# Analyze special characters usage
special_char_features = ['IsDomainIP', 'NoOfLettersInURL', 'NoOfDegitsInURL', 
                         'NoOfDash', 'NoOfAt', 'NoOfQMark', 'NoOfAnd', 'NoOfEqual',
                         'NoOfUnderscore', 'NoOfPercent', 'NoOfSlash', 'NoOfHash', 'IsHTTPS']

fig, axes = plt.subplots(3, 1, figsize=(16, 12))
axes = axes.flatten()

for idx, feature in enumerate(special_char_features[:9]):
    if feature in df.columns:
        phishing = df[df['label']==0][feature]
        legitimate = df[df['label']==1][feature]
        
        axes[idx].hist(phishing, bins=20, alpha=0.6, label='Phishing', color='#e74c3c')
        axes[idx].hist(legitimate, bins=20, alpha=0.6, label='Legitimate', color='#2ecc71')
        axes[idx].set_title(feature)
        axes[idx].set_xlabel('Count')
        axes[idx].set_ylabel('Frequency')
        axes[idx].legend()
        axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nSpecial Characters Statistics (Mean values):")
available_features = [f for f in special_char_features if f in df.columns]
print(df.groupby('label')[available_features].mean())
```

### 6. Correlation Heatmap

```python
# Select numeric features for correlation analysis
numeric_features = ['URLLength', 'DomainLength', 'IsDomainIP', 'TLDLength', 
                    'NoOfSubDomain', 'HasObfuscation', 'NoOfObfuscatedChar',
                    'ObfuscationRatio', 'NoOfLettersInURL', 'NoOfDegitsInURL',
                    'NoOfEqualsInURL', 'NoOfQMarkInURL', 'NoOfAmpersandInURL',
                    'NoOfOtherSpecialCharsInURL', 'SpacialCharRatioInURL',
                    'IsHTTPS', 'LineOfCode', 'LargestLineLength', 'label']

# Filter only columns that exist
numeric_features = [col for col in numeric_features if col in df.columns]

# Calculate correlation matrix
corr_matrix = df[numeric_features].corr()

# Plot heatmap
plt.figure(figsize=(16, 14))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Show features most correlated with label
print("\nTop Features Correlated with Label (Phishing Detection):")
label_corr = corr_matrix['label'].abs().sort_values(ascending=False)
print(label_corr.head(20))
```

### 7. Key Insights Summary

```python
# Generate comprehensive insights
print("=" * 80)
print("KEY FINDINGS FROM EDA".center(80))
print("=" * 80)

# 1. Dataset balance
label_counts = df['label'].value_counts()
balance_ratio = label_counts[1] / label_counts[0]
print(f"\n1. DATASET BALANCE:")
print(f"   • Total samples: {len(df):,}")
print(f"   • Phishing: {label_counts[0]:,} ({label_counts[0]/len(df)*100:.1f}%)")
print(f"   • Legitimate: {label_counts[1]:,} ({label_counts[1]/len(df)*100:.1f}%)")
print(f"   • Balance ratio: {balance_ratio:.2f}:1 {'⚠️ IMBALANCED' if balance_ratio > 1.5 or balance_ratio < 0.67 else '✓ BALANCED'}")

# 2. URL Length insights
print(f"\n2. URL LENGTH PATTERNS:")
print(f"   • Phishing URLs: {df[df['label']==0]['URLLength'].mean():.1f} chars (avg)")
print(f"   • Legitimate URLs: {df[df['label']==1]['URLLength'].mean():.1f} chars (avg)")
length_diff = abs(df[df['label']==0]['URLLength'].mean() - df[df['label']==1]['URLLength'].mean())
print(f"   • Difference: {length_diff:.1f} chars {'📊 SIGNIFICANT' if length_diff > 10 else ''}")

# 3. HTTPS usage
https_phish = (df[df['label']==0]['IsHTTPS'].sum() / len(df[df['label']==0])) * 100
https_legit = (df[df['label']==1]['IsHTTPS'].sum() / len(df[df['label']==1])) * 100
print(f"\n3. HTTPS USAGE:")
print(f"   • Phishing: {https_phish:.1f}% use HTTPS 🚨")
print(f"   • Legitimate: {https_legit:.1f}% use HTTPS")
print(f"   • Note: HTTPS doesn't guarantee legitimacy!")

# 4. Domain characteristics
print(f"\n4. DOMAIN CHARACTERISTICS:")
print(f"   • Phishing domain length: {df[df['label']==0]['DomainLength'].mean():.1f} chars")
print(f"   • Legitimate domain length: {df[df['label']==1]['DomainLength'].mean():.1f} chars")
print(f"   • Phishing subdomains: {df[df['label']==0]['NoOfSubDomain'].mean():.2f} (avg)")
print(f"   • Legitimate subdomains: {df[df['label']==1]['NoOfSubDomain'].mean():.2f} (avg)")

# 5. Special characters
if 'NoOfDash' in df.columns:
    print(f"\n5. SPECIAL CHARACTERS (Phishing vs Legitimate):")
    print(f"   • Dashes: {df[df['label']==0]['NoOfDash'].mean():.2f} vs {df[df['label']==1]['NoOfDash'].mean():.2f}")
if 'NoOfSlash' in df.columns:
    print(f"   • Slashes: {df[df['label']==0]['NoOfSlash'].mean():.2f} vs {df[df['label']==1]['NoOfSlash'].mean():.2f}")
if 'NoOfQMark' in df.columns:
    print(f"   • Question marks: {df[df['label']==0]['NoOfQMark'].mean():.2f} vs {df[df['label']==1]['NoOfQMark'].mean():.2f}")

# 6. Obfuscation
if 'HasObfuscation' in df.columns:
    obf_phish = (df[df['label']==0]['HasObfuscation'].sum() / len(df[df['label']==0])) * 100
    obf_legit = (df[df['label']==1]['HasObfuscation'].sum() / len(df[df['label']==1])) * 100
    print(f"\n6. OBFUSCATION:")
    print(f"   • Phishing with obfuscation: {obf_phish:.1f}%")
    print(f"   • Legitimate with obfuscation: {obf_legit:.1f}%")

print("\n" + "=" * 80)
```
